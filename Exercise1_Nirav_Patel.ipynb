```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('https://raw.githubusercontent.com/niravpatidar37/EPL/main/final_dataset.csv')
```

```python
# Calculate average goals scored by home teams and away teams
avg_home_goals = df['FTHG'].mean()
avg_away_goals = df['FTAG'].mean()

print(f"Average goals scored by home teams: {avg_home_goals:.2f}")
print(f"Average goals scored by away teams: {avg_away_goals:.2f}")
```

```python
# Identify teams with the highest average goals scored
team_with_max_home_goals = df.groupby('HomeTeam')['FTHG'].mean().idxmax()
team_with_max_away_goals = df.groupby('AwayTeam')['FTAG'].mean().idxmax()


print(f"Team with the highest average home goals: {team_with_max_home_goals}")
print(f"Team with the highest average away goals: {team_with_max_away_goals}")
```

```python
# Analyze distribution of wins, losses, and draws for home teams and away teams
home_team_results = df['FTR'].value_counts()
away_team_results = df['FTR'].replace({'H': 'A', 'A': 'H'}).value_counts()

print("\nDistribution of results for home teams:")
print(home_team_results)
print("\nDistribution of results for away teams:")
print(away_team_results)
```

```python
# Calculate total points for each team
df['TotalPoints'] = df['HTP'] + df['ATP']
team_points = df.groupby('HomeTeam')['TotalPoints'].sum()

print("\nTotal points earned by teams:")
print(team_points)
```

```python
# Identify top-performing teams based on total points
top_teams = team_points.nlargest(5)
print("\nTop-performing teams based on total points:")
print(top_teams)
```

```python
# Analyze relationship between goals scored and team performance
goals_vs_points = df.groupby('HomeTeam')[['HTGS', 'ATGS', 'TotalPoints']].mean()
print("\nRelationship between goals scored and team performance:")
print(goals_vs_points)
```

```python
# Analyze team form and streaks
avg_form_points_home = df['HTFormPtsStr'].apply(lambda x: sum(int(val) for val in x.split(',') if val.isdigit())).mean()
avg_form_points_away = df['ATFormPtsStr'].apply(lambda x: sum(int(val) for val in x.split(',') if val.isdigit())).mean()
```

```python
# Calculate goal difference for each team
df['GoalDifference'] = df['HTGD'] + df['ATGD']
team_goal_difference = df.groupby('HomeTeam')['GoalDifference'].sum()

print("\nGoal difference for each team:")
print(team_goal_difference)
```

```python
# Identify teams with the highest positive or negative goal differences
team_with_max_goal_diff = team_goal_difference.idxmax()
team_with_min_goal_diff = team_goal_difference.idxmin()

print(f"\nTeam with the highest positive goal difference: {team_with_max_goal_diff}")
print(f"Team with the highest negative goal difference: {team_with_min_goal_diff}")
```

```python
# Analyze relationship between goal difference and points earned
goal_diff_vs_points = df.groupby('HomeTeam')[['GoalDifference', 'TotalPoints']].mean()
print("\nRelationship between goal difference and points earned:")
print(goal_diff_vs_points)
```

```python
# Analyze trends across matchweeks
avg_goals_per_matchweek = df.groupby('MW')[['FTHG', 'FTAG']].mean()
```

```python
# Set the style for seaborn plots
sns.set(style='darkgrid')

# Histogram of Home Team Goals
plt.hist(df['FTHG'])
plt.xlabel('Home Team Goals')
plt.ylabel('Frequency')
plt.title('Distribution of Home Team Goals')
plt.show()
```

```python
# Histogram of Away Team Goals
plt.hist(df['FTAG'])
plt.xlabel('Away Team Goals')
plt.ylabel('Frequency')
plt.title('Distribution of Away Team Goals')
plt.show()
```

```python
# Scatter plot of Home Team Goals vs Away Team Goals
plt.scatter(df['FTHG'], df['FTAG'])
plt.xlabel('Home Team Goals')
plt.ylabel('Away Team Goals')
plt.title('Home Team Goals vs Away Team Goals')
plt.show()
```

```python
# Create a bar plot of the distribution of results for home teams
plt.figure(figsize=(8, 6))
sns.countplot(x='FTR', data=df)
plt.xlabel('Result')
plt.ylabel('Count')
plt.title('Distribution of Results for Home Teams')
plt.show()
```

```python
# Create a bar plot of the distribution of results for away teams
plt.figure(figsize=(8, 6))
sns.countplot(x=df['FTR'].replace({'H': 'A', 'A': 'H'}))
plt.xlabel('Result')
plt.ylabel('Count')
plt.title('Distribution of Results for Away Teams')
plt.show()
```

```python
# Create a bar plot of top-performing teams based on total points
plt.figure(figsize=(10, 6))
sns.barplot(x=top_teams.index, y=top_teams.values)
plt.xlabel('Team')
plt.ylabel('Total Points')
plt.title('Top-performing Teams based on Total Points')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a scatter plot of goals scored vs. total points for home teams
plt.figure(figsize=(8, 6))
sns.scatterplot(x='HTGS', y='TotalPoints', data=df)
plt.xlabel('Home Team Goals Scored')
plt.ylabel('Total Points')
plt.title('Goals Scored vs. Total Points for Home Teams')
plt.show()
```

```python
# Create a scatter plot of goals scored vs. total points for away teams
plt.figure(figsize=(8, 6))
sns.scatterplot(x='ATGS', y='TotalPoints', data=df)
plt.xlabel('Away Team Goals Scored')
plt.ylabel('Total Points')
plt.title('Goals Scored vs. Total Points for Away Teams')
plt.show()
```

```python
# Create a box plot of goal difference for each team
plt.figure(figsize=(10, 6))
sns.boxplot(x=team_goal_difference)
plt.xlabel('Team')
plt.ylabel('Goal Difference')
plt.title('Goal Difference for Each Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a line plot of average goals scored per matchweek
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_goals_per_matchweek)
plt.xlabel('Matchweek')
plt.ylabel('Average Goals Scored')
plt.title('Average Goals Scored per Matchweek')
plt.legend(['Home Team Goals', 'Away Team Goals'])
plt.show()
```
